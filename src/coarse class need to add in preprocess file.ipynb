```python
def coarse_classing(self, df):

    # home ownership
    df['home_ownership_grp'] = df['home_ownership'].replace({
        'ANY': 'RENT_ANY',
        'RENT': 'RENT_ANY',
        'OWN': 'OWN',
        'MORTGAGE': 'MORTGAGE',
        'NONE': 'OTHER_NONE',
        'OTHER': 'OTHER_NONE'
    })

    # addr_state (using WoE thresholds)
    state_map = self.state_woe_map  # store this once (see below)

    def group_state(woe):
        if woe < -0.3:
            return 'very_high_risk'
        elif woe < 0:
            return 'high_risk'
        elif woe < 0.2:
            return 'medium_risk'
        else:
            return 'low_risk'

    df['addr_state_grp'] = df['addr_state'].map(state_map).apply(group_state)

    return df
```
